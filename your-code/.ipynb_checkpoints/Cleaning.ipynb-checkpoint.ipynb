```python
db.Age=db.Age.str.replace(r'^.een$',"18")
db.Age=db.Age.str.replace(r'^.eens$',"18")
db.Age=db.Age.str.replace(r'^.oun.$', "25")
db.Age=db.Age.str.replace(r'^.young''.$',"25")
db.Age=db.Age.str.replace(r'^.months$',"1")
db.Age=db.Age.str.replace(r'(^.*onth.*$)', "1")
db.Age=db.Age.str.replace(r'(^.*dul.*$)', "40")
db.Age=db.Age.str.replace(r'(^.*lderl.*$)', "70")
db.Age=db.Age.str.replace(r'(^.*M..*$)', "NaN")
db.Age=db.Age.str.replace(r'(^.*middle-age.*$)',"35")
db.Age=db.Age.str.replace(r'(.*)\\sor.*', r'\\1')
db.Age=db.Age.str.replace(r'(^.*GREEN*$)', "Nan")
db.Age=db.Age.str.replace(r'^\\s\\s.*', r"NaN")
db.Age=db.Age.str.replace(r'(^.*Both.*$)',"11")
db.Age=db.Age.str.replace(r'(^.*mid.*$)',"35")
db.Age=db.Age.str.replace(r'(^.*or.*$)',"10")
db.Age=db.Age.str.replace(r'(^.*to.*$)',"30")
db.Age=db.Age.str.replace(r'(^.*F.*$)',"Nan")
db.Age=db.Age.str.replace(r'(^.*X.*$)',"Nan")
db.Age=db.Age.str.replace(r'(^.*Ca.*$)',"33")
db.Age=db.Age.str.replace(r'(^.*&.*$)',"NaN")
db.Age=db.Age.str.replace(r'(^.*43.*$)',"43")
db.Age=db.Age.str.replace(r'(^.*40.*$)',"40")
db.Age=db.Age.str.replace(r'(^.*60.*$)',"60")
db.Age=db.Age.str.replace(r'(^.*30.*$)',"30")
db.Age=db.Age.str.replace(r'(^.*20.*$)',"20")
db.Age=db.Age.str.replace(r'^\s.*', r"NaN")
db.Age=db.Age.str.replace(r'(^.*50.*$)',"30")
db.Age=db.Age.str.replace(r'(^.*2½.*$)',"3")
db.Age=db.Age.str.replace(r'(^.*6½.*$)',"7")
db.Age=db.Age.str.replace("nan", "NaN")
db.Age=db.Age.str.replace("Nan", "NaN")
db.Age=db.Age.str.replace("NaN", "0")
```
